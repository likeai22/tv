# Задание 1

Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.


```python
data = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]
```

```python
import numpy as np
n = len(data)
data_sort = np.msort(data)

mx = sum(data) / n
s = sum(([(i - sum(data) / n) ** 2 for i in data])) / len(data)
s0 = sum(([(i - sum(data) / n) ** 2 for i in data])) / (len(data) - 1)
sigma = np.sqrt(s)
sigma0 = np.sqrt(s0)

print('Исходная выборка:', data)
print('Объем выборки:', n)
print('Среднее:', mx)
print('Дисперсия генеральной совокупности:', s)
print('Дисперсия выборки:', s0)
print('Среднеквадратическое отклонение генеральной совокупности', sigma)
print('Среднеквадратическое отклонение выборки', sigma0)
```

Среднее квадратичное отклонение с помощью **std**:

```python
np.std(data, ddof=0)
```

```python
np.std(data, ddof=1)
```

С помощью **var** также можно сделать оценки дисперсии:

```python
np.var(data, ddof=0)
```

```python
np.var(data, ddof=1)
```

## Задание 2
В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

Событие $X -$ три мяча белые

| Событие | Описание |
|------|------|
|A| из первого 0 белых, из второго 3 белых|
|B| из первого 1 белый, из второго 2 белых|
|C| из первого 2 белых, из второго 1 белый|

$$ P(A) = \frac{C^0_5 \cdot C^2_3}{C^2_8} \cdot \frac{C^3_5 \cdot C^1_7}{C^4_{12}} = 0.0152 $$
$$ P(B) = \frac{C^1_5 \cdot C^1_3}{C^2_8} \cdot \frac{C^2_5 \cdot C^2_7}{C^4_{12}} = 0.2273 $$
$$ P(C) = \frac{C^2_5 \cdot C^0_3}{C^2_8} \cdot \frac{C^1_5 \cdot C^3_7}{C^4_{12}} = 0.1263 $$

$$ P(X)=P(A) + P(B) + P(C) = \\ = 0.0152 + 0.2273 + 0.1263 = 0.3687 $$

```python
from math import factorial as f


def c(k, n):
    return f(n) / (f(k) * f(n - k))
```

```python
P_a = c(0, 5) * c(2, 3) / c(2, 8) * c(3, 5) * c(1, 7) / c(4, 12)
P_b = c(1, 5) * c(1, 3) / c(2, 8) * c(2, 5) * c(2, 7) / c(4, 12)
P_c = c(2, 5) * c(0, 3) / c(2, 8) * c(1, 5) * c(3, 7) / c(4, 12)

print(f'''Вероятности событий:
 A: {P_a:.4f},
  B: {P_b:.4f},
   C: {P_c:.4f},
    X: {P_a + P_b + P_c:.4f}''')
```

## 3. Задание 3

На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a) первым спортсменом б) вторым спортсменом в) третьим спортсменом.

Событие ${B-}$ спортсмен выстрелил. Вероятность выстрела любым спортсменом $P(B_i) = \frac 1 3$

Событие ${A-}$ спортсмен попал. При условии, что спортсмен выстрелил, условные вероятности попаданий:

$P(A|B_{1}) = 0.9, P(A|B_{2}) = 0.8, P(A|B_{3}) = 0.6$

Найдем полную вероятность:
$$P(A)=P(B_{1})\cdot P(A|B_{1})+P(B_{2})\cdot P(A|B_{2})+ \dots +P(B_{n})\cdot P(A|B_{n})$$
$$P(A) = 1/3 \cdot (0.9 + 0.8 + 0.6) = 0.767$$

Согласно теореме Байеса:

a) вероятность того, что выстрел произведен первым спортсменом
$$P(B_1\:|\:A) = \frac{P(B_1) \cdot P(A\:|\:B_1)}{P(A)} = \frac{1}{3} \cdot \frac{0.9}{0.767} = 0.391 $$

б) вторым спортсменом
$$P(B_2\:|\:A) = \frac{P(B_2) \cdot P(A\:|\:B_2)}{P(A)} = \frac{1}{3} \cdot \frac{0.8}{0.7666} = 0.348 $$

в) третьим спортсменом
$$P(B_3\:|\:A) = \frac{P(B_3) \cdot P(A\:|\:B_3)}{P(A)} = \frac{1}{3} \cdot \frac{0.6}{0.7666} = 0.261 $$

Проверка: $0,391 + 0,348 + 0,261 = 1$

## Задание 4
В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a) на факультете A б) на факультете B в) на факультете C?

Обозначим через событие ${A}$ — студент сдал сессию

Событием ${B_1}$ — сдавший сессию студент учился на факультета ${A}$, ${B_2}$ — факультете ${B}$, ${B_3}$ — факультете ${C}$.

$P(B_1) = X$, $P(B_2) = X$, $P(B_3) = 2X$ и

$P(B_1) + P(B_2) + P(B_3) = 1$

В данном случае, вероятность быть студентом любого из факультетов можно оставить $X$, нас она не интересует по условию и потом сократится при расчёте условных вероятностей.

Вероятность сдать сессию для студентов каждого факультета  равны: $P(A|B_{1}) = 0.8$, $P(A|B_{2}) = 0.7$, $P(A|B_{3}) = 0.9$

Заметим, что случайное событие ${A}$ может произойти только лишь с одной из гипотез. Сформулированные гипотезы попарно несовместны и образуют полную группу событий. Значит можно воспользоваться формулой полной вероятности.

$$P(A) = X\cdot (0.8 + 0.7 + 2 \cdot 0.9) = 3.3\cdot X$$

Найдем условные вероятности события ${A}$ при условии реализации каждой из гипотез.

Вероятность, что студент, который сдал первую сессию, учится:

a) на факультете A

$$P(B_1\:|\:A) = \frac{P(B_1) \cdot P(A\:|\:B_1)}{P(A)} = \frac{X \cdot 0.8}{3.3 \cdot X} = 0.242 $$

б) на факультете B
$$P(B_2\:|\:A) = \frac{P(B_2) \cdot P(A\:|\:B_2)}{P(A)} = \frac{X \cdot 0.7}{3.3 \cdot X} = 0.212 $$

в) на факультете С
$$P(B_3\:|\:A) = \frac{P(B_3) \cdot P(A\:|\:B_3)}{P(A)} = \frac{2X \cdot 0.9}{3.3 \cdot X} = 0.546 $$

## Задание 5
Устройство состоит из трех деталей. Для первой детали вероятность выйти из строя в первый месяц равна 0.1, для второй - 0.2, для третьей - 0.25. Какова вероятность того, что в первый месяц выйдут из строя: а) все детали б) только две детали в) хотя бы одна деталь г) от одной до двух деталей?

| Событие | Вероятность |
|------|------|
|A| 0.1, вышла из строя 1 деталь|
|B| 0.2, вышла из строя 2 деталь|
|C| 0.25, вышла из строя 3 деталь|

Возможные случаи можно разделить на группы

| Событие | Из строя вышли детали |
|------|------|
|X| все детали|
|Y| две детали|
|Z| хотя бы одна деталь|
|Q| от одной до двух деталей|

$P(X) + P(Y) + P(Z) + P(Q) = 1$

Выразим события формулами:

$X=A\cdot B\cdot C=0.1\cdot 0.2\cdot 0.25 = 0.005$

$Y=A\cdot B\cdot \bar{C}+A\cdot \bar{B}\cdot C+\bar{A}\cdot B\cdot C = 0.1 \cdot 0.2 \cdot (1 - 0.25) + 0.1 \cdot (1 - 0.2) \cdot 0.25 +\\+ (1 - 0.1) \cdot 0.2 \cdot 0.25 = 0.015 + 0.02 + 0.045 = 0.08$

$Z=\overline{\bar{A}\cdot \bar{B}\cdot \bar{C}}=1 - ((1 - 0.1) \cdot (1 - 0.2) \cdot (1 - 0.25)) = 1 - 0.54 = 0.46 $

$Q=1 - (P(X) + P(Y) + P(Z)) = 1 - (0.005 + 0.08 + 0.46) = 0.455$
